```python
import pandas as pd
```

```python
origin_df = pd.read_csv('data/WID_Data_Metadata/WID_Data_20112025-013227.csv', sep=';', skiprows=1, encoding='utf-8')
origin_df['mean_of_gini_coefficient'] = origin_df.iloc[:, 2:13].mean(axis=1)
origin_df['mean_of_carbon_emissions'] = origin_df.iloc[:, 13:24].mean(axis=1)
origin_df.columns = origin_df.columns.str.rstrip()
mean_df=origin_df[['Country', 'mean_of_gini_coefficient', 'mean_of_carbon_emissions']]

print('df shape:', mean_df.shape)
print('Country number of unique:', mean_df['Country'].nunique())
print(mean_df.describe())
numeric_cols = mean_df.select_dtypes(include='number').columns
mean_df[numeric_cols].corr()
```
